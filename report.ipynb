```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

M_IA = 130

# https://pure.tue.nl/ws/files/2488354/200510540.pdf
# eq. 4-3   
def w(cld):
    return cld[:,1]*(cld[:,0]*M_IA)**2

def logM(cld):
    return np.log10(cld[:,0]*M_IA)

# powtórzenie 1
aD1h = np.loadtxt('01/D.3.60000000000000e+03.cld', skiprows=3)
aD2h = np.loadtxt('01/D.7.20000000000000e+03.cld', skiprows=3)
aD4h = np.loadtxt('01/D.1.44000000000000e+04.cld', skiprows=3)
aD8h = np.loadtxt('01/D.2.88000000000000e+04.cld', skiprows=3)
aD16h = np.loadtxt('01/D.5.76000000000000e+04.cld', skiprows=3)
aD24h = np.loadtxt('01/D.8.64000000000000e+04.cld', skiprows=3)

# powtórzenie 2
bD1h = np.loadtxt('02/D.3.60000000000000e+03.cld', skiprows=3)
bD2h = np.loadtxt('02/D.7.20000000000000e+03.cld', skiprows=3)
bD4h = np.loadtxt('02/D.1.44000000000000e+04.cld', skiprows=3)
bD8h = np.loadtxt('02/D.2.88000000000000e+04.cld', skiprows=3)
bD16h = np.loadtxt('02/D.5.76000000000000e+04.cld', skiprows=3)
bD24h = np.loadtxt('02/D.8.64000000000000e+04.cld', skiprows=3)

# powtórzenie 3
cD1h = np.loadtxt('03/D.3.60000000000000e+03.cld', skiprows=3)
cD2h = np.loadtxt('03/D.7.20000000000000e+03.cld', skiprows=3)
cD4h = np.loadtxt('03/D.1.44000000000000e+04.cld', skiprows=3)
cD8h = np.loadtxt('03/D.2.88000000000000e+04.cld', skiprows=3)
cD16h = np.loadtxt('03/D.5.76000000000000e+04.cld', skiprows=3)
cD24h = np.loadtxt('03/D.8.64000000000000e+04.cld', skiprows=3)




plt.xlabel('logM')
plt.ylabel('dw/dlogM')

plt.plot(logM(aD1h), w(aD1h), label='1h')
#plt.plot(logM(bD1h), w(bD1h))
#plt.plot(logM(cD1h), w(cD1h))

plt.plot(logM(aD2h), w(aD2h), label='2h')
plt.plot(logM(aD4h), w(aD4h), label='4h')
plt.plot(logM(aD8h), w(aD8h), label='8h')
plt.plot(logM(aD16h), w(aD16h), label='16h')
plt.plot(logM(aD24h), w(aD24h), label='24h')
plt.legend()
```
